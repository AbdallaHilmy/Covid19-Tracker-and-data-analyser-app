```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r'Downloads\Coronavirus_17Apr.csv')
```

```python
df_fill = df.fillna("")
```

```python
df_num = df_fill.apply(lambda x: pd.factorize(x)[0])+ 1
```

```python
corr = df_num.corr()
ax = sns.heatmap(
corr,
vmin=-1, vmax=1, center=0,
cmap=sns.diverging_palette(20, 220, n=200),
square=True
)
ax.set_xticklabels(
ax.get_xticklabels(),
rotation=45,
horizontalalignment='right'
);
```
